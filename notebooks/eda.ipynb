# Exploratory Data Analysis (EDA) on Concrete Compressive Strength Dataset
In this notebook, we'll explore the dataset to understand the features, check for missing values, and visualize relationships between variables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
plt.style.use('ggplot')
sns.set_theme(style='whitegrid')
```

```python
# Load the dataset
# Assuming the notebook is in the 'notebooks' directory and the dataset is in the 'data' directory
data_path = '../data/Concrete_Data.xls'
df = pd.read_excel(data_path)

# Display the first few rows
df.head()
```

### Data Information and Missing Values

```python
print("Dataset Info:")
df.info()
```

```python
print("Missing Values:")
df.isnull().sum()
```

### Summary Statistics

```python
df.describe()
```

### Data Visualization

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Concrete Features')
plt.show()
```

```python
# Distribution of the target variable (Concrete compressive strength)
target_col = df.columns[-1]  # Assuming the last column is the target
plt.figure(figsize=(8, 6))
sns.histplot(df[target_col], kde=True, bins=30, color='blue')
plt.title(f'Distribution of {target_col}')
plt.xlabel('Compressive Strength (MPa)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pairplot to see relationships between variables
# (This might take a moment to render)
sns.pairplot(df, diag_kind='kde')
plt.show()
```
